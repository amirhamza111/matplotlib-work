```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 50
mu = 3

x = np.random.normal(mu,1,size=n)
y = np.random.normal(mu,1,size=n)
z = np.random.normal(mu,1,size=n)
```

```python
plt.rcParams["figure.figsize"] = (8,6)
```

```python
# Scatter Plot
ax = plt.axes(projection="3d")
ax.scatter3D(x,y,z,s=100)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

ax.view_init(45,215);
```

```python
#3D line plot
omega = 2

z_line = np.linspace(0,10,100)
x_line = np.cos(omega*z_line)
y_line = np.sin(omega*z_line)

ax = plt.axes(projection="3d")
ax.plot3D(x_line,y_line,z_line,lw=10);
```

```python
# 3d Surface plot

def function_z(x,y):
  return 50-(x**2+y**2)

N = 50

x_values = np.linspace(-5,5,N)
y_values = np.linspace(-5,5,N)

X , Y = np.meshgrid(x_values,y_values)

# plt.scatter(X,Y)

Z = function_z(X,Y)
# Z.shape

ax = plt.axes(projection="3d")
ax.plot_surface(X,Y,Z);
```

```python
# 3D wireframe
N = 10

x_values = np.linspace(-5,5,N)
y_values = np.linspace(-5,5,N)

X , Y = np.meshgrid(x_values,y_values)

# plt.scatter(X,Y)

Z = function_z(X,Y)
# Z.shape

ax = plt.axes(projection="3d")
ax.plot_wireframe(X,Y,Z,lw=3);
```
